```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
import matplotlib

data = pd.read_csv("data2/house_raw.csv")

X = data[data.columns[0:5]]
y = data[['house_value']]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

scaler_minmax=MinMaxScaler()
scaler_minmax.fit(X_train)

X_scaled_minxmax_train = scaler_minmax.transform(X_train)
X_scaled_minmax_test = scaler_minmax.transform(X_test) 

model=LinearRegression()
model.fit(X_scaled_minxmax_train, y_train)

pred_train=model.predict(X_scaled_minmax_test)
model.score(X_scaled_minmax_test, y_test)

# data.hist(bins=50, figsize=(50,50))

data_bedroom = data[data['bedrooms'] < 0.6]

data_bedroom2 = data[data['bedrooms'] >= 0.6]

data_households = data[data['households'] < 10]

data_households.hist(bins=100, figsize=(20,15))

data_rooms = data[data['rooms'] < 20]
data_rooms.hist(bins=100, figsize = (20,15))
```
